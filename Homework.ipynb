```python
import random
import datetime
import xpinyin

p = xpinyin.Pinyin()

print('游戏规则：\n根据提问的省份名称写出其对应的汉语拼音并用数字表示其音调\n'
      '例如：提问：安徽\n      回答：an1 hui1\n'
      '你完成测试的用时将会计入成绩榜，来冲击你的最佳成绩吧！')
while 1:  # 避免输入错误而导致程序无法运行的循环
    print('按Y开始')
    start = input()
    if start == 'Y':  # 判断使用者是否输入为Y
        scores = []  # 引入成绩榜列表
        number = 0
        while 1:  # 开始游戏的循环
            pro = ['河北', '山西', '辽宁', '吉林',
                   '黑龙江', '江苏', '浙江', '安徽',
                   '福建', '江西', '山东', '河南', '台湾',
                   '湖北', '湖南', '广东', '海南', '四川',
                   '贵州', '云南', '陕西', '甘肃', '青海',
                   '内蒙古', '广西', '宁夏', '新疆',
                   '北京', '上海', '天津', '重庆', '香港', '澳门']  # 引入包含所有省份的列表
            times = 0  # 初始化完成次数
            num = [-1, -1, -1]  # 引入数组容纳三个随机数
            for i in range(3):  # 开始三次测试
                mytime1 = datetime.datetime.now()  # 生成开始时间
                while 1:  # 进行循环来避免生成重复的随机数
                    num[i] = random.randint(0, 32)  # 生成测试所用随机数
                    for z in range(i):  # 检查是否有与前面相同随机数
                        if num[z] == num[i]:  # 判断是否存在
                            num[i] = -1  # 存在相同则又初始化
                    if num[i] != -1:
                        break
                print('提问：'+pro[num[i]])
                answer0 = p.get_pinyin(pro[num[i]], ' ', tone_marks='numbers')  # 得到正确答案
                answer = input('请输入你的答案：')
                if answer == answer0:
                    mytime2 = datetime.datetime.now()  # 生成结束时间
                    if i == 0:
                        usetime = mytime2 - mytime1  # 计算第一次用时
                    else:
                        usetime = usetime + (mytime2 - mytime1)  # 累计用时
                    times = times + 1  # 累计完成次数
                    print('回答正确！请看下一题')
                else:
                    print('很可惜，回答错误。正确答案是'+answer0)
            if times == 3:
                print('恭喜你成功完成三次测试!\n本次测试用时'+str(usetime))  # 输出全部完成所用时间
                if usetime in scores:  # 判断是否有成绩重复
                    for x in range(number):  # 用循环输出成绩榜
                        print('第'+str(x+1)+'名：'+str(scores[x]))
                else:
                    number = number + 1  # 累计成绩数目
                    scores.append(usetime)  # 插入新成绩
                    scores.sort()  # 整理成绩榜
                    if usetime == scores[0]:
                        print('恭喜创造新纪录')
                    for y in range(number):  # 循环输出成绩榜
                        print('第'+str(y+1)+'名：', scores[y])
            else:
                print('你成功完成'+str(times)+'次测试，很遗憾，看来要再努力一点呢')  # 输出完成次数
            while 1:  # 避免使用者输入错误而进行的循环
                print('继续测试请输入again，结束测试请输入quit')
                choose = input()
                if choose == 'again' or choose == 'quit':  # 判断输入的单词是否正确
                    break
                else:
                    continue
            if choose == 'again':  # 判断使用者是否继续
                print('开始下一次测试咯！')
                continue
            else:
                break  # 跳出测试循环
        else:
            continue
        if choose == 'quit':
            break  # 结束测试
```

```python
import jieba
import jieba.posseg as pseg

'''
用jieba库完成词频统计
'''
file = open('水浒传.txt', 'r', encoding = 'utf-8')  # 打开txt文件
content = file.read()  # 读取
file.close()

# 设置忽略词语列表
ignore_list = ['\n', '二人', '一个', '来到', '哥哥',
               '人马', '你们', '我们', '好汉', '知府',
               '什么', '头领', '他们', '银子', '梁山',
               '两个', '山寨', '众人', '出来', '大汉',
               '就是', '三人', '上山', '怎么', '次日',
               '和尚', '兄弟', '众人', '这里', '公人',
               '书信', '慌忙', '汉子', '只好', '回来',
               '如何', '下山', '回到', '不是', '酒店',
               '赶来', '过来', '一阵', '几个', '收拾',
               '起来', '喽啰', '一声', '不知', '不敢',
               '却是', '梁山泊', '哪里', '吃酒', '几天',
               '庄客', '正是', '那人', '回合', '人称',
               '一把', '酒保', '听说', '女人', '一条',
               '施恩', '师父', '太公', '一面']

word_list = jieba.lcut(content)

word_dict = {}

# 进行循环统计词频
for word in word_list:
    # 忽略忽略列表词语与长度为1的词语
    if word in ignore_list or len(word) == 1:
        continue

    # 合并相同人名
    if word == '智深' or word == '鲁达':
        word = '鲁智深'
    elif word == '教头':
        word = '林冲'
    else:
        pass
    
    if word in word_dict.keys():
        word_dict[word] = word_dict[word] + 1  #如果存在则value加一
    else:
        word_dict[word] = 1  #不存在则令其value为1

# 将字典转化为列表排序
items_list = list(word_dict.items())
items_list.sort(key=lambda x: x[1], reverse=True)

# 统计总词语数目
total_num = len(items_list)
print('经统计，共有' + str(total_num) + '个不同的词')

# 根据用户需求，打印排名前列的词，同时把统计结果存入文件
num = int(input('您想查看前多少个人物？：'))

# 写入csv文件
result_file = open('result1.csv', 'w')
result_file.write('人物,出现次数\n')
for i in range(num):
    word, cnt = items_list[i]
    print(word + ' ' + str(cnt))
    result_file.write(word + ',' + str(cnt) + '\n')
result_file.close()

print('已成功写入result1.csv')
print()

# %%

file = open('水浒传.txt', 'r', encoding = 'utf-8')  # 打开txt文件
content = file.read()  # 读取文件
file.close()

words = pseg.cut(content)  # 分词

word_dict = {}  #用于统计词频的字典

print()
print('正在统计所有词中的人物名词……')
count = 0  # 用于统计已经处理多少词语

for one in words: 
    # 便于编写程序，用首字母代替
    w = one.word
    f = one.flag

    if len(w) == 1:  # 忽略单字
        continue

    # 综合相同人名
    if w == '智深' or w == '鲁达':
        w = '鲁智深'
    elif w == '教头':
        w = '林冲'
    elif w == '及时雨':
        w == '宋江'
    elif w == '黑旋风':
        w == '李逵'
    else:
        pass

    # 统计词频
    if 'nr' in f: # 如果该词的词性中包含'nr'，即这是个人物名词
        if w in word_dict.keys(): # 如果该词已经在词典中，则value加一
            word_dict[w] = word_dict[w] + 1
        else: # 如果该词不在词典中，则使其value为1
            word_dict[w] = 1

    # 打印进度
    count = count + 1
    count_quo = int(count/1000)
    count_mod = count % 1000 # 取模，即做除法得到的余数
    if count_mod == 0: # 每逢整千的数，打印一次进度
        print('\r已处理词数：' + '-'*count_quo + '> '\
              + str(count_quo) + '千', end='')  # 自行刷新的进度条

print()
print('人物名词统计完成')

# 将字典转化为列表排序
items_list = list(word_dict.items())
items_list.sort(key=lambda x: x[1], reverse=True)

# 统计总共的任务名词
total_num = len(items_list)
print('经统计，共有' + str(total_num) + '个不同的人物名词')

# 根据用户需求，打印排名前列的词，同时把统计结果存入文件
num = int(input('您想查看前多少个人物？'))

# 写入csv文件
result_file = open('result2.csv', 'w')
result_file.write('人物,出现次数\n')
for i in range(num):
    word, cnt = items_list[i]
    print(word + ' ' + str(cnt))
    result_file.write(word + ',' + str(cnt) + '\n')
result_file.close()

print('已成功写入result2.csv')

# %%

file = open('水浒传.txt', 'r', encoding='utf-8')
line_list = file.readlines() # 返回列表，每一行（段落）是列表的一个元素
file.close()

# 准备工作
line_name_list = []  # 用于统计每个段落出现的人物列表的列表
name_cnt_dict = {}  # 用于统计人物出现次数

print()
print('正在统计所有词中的人物名词……')
count = 0  # 用于统计已经处理的词汇数目

for line in line_list:
    word_gen = pseg.cut(line)  # 处理每一段的词语
    line_name_list.append([])  # 添加该段落的人物列表
    for one in word_gen:  # 用首字母代替便于编程
        w = one.word
        f = one.flag

        if len(w) == 1:  # 忽略长度为1的词汇
            continue
        
        # 综合相同人名
        if w == '智深' or w == '鲁达':
            w = '鲁智深'
        elif w == '教头':
            w = '林冲'
        elif w == '及时雨':
            w == '宋江'
        elif w == '黑旋风':
            w == '李逵'
        else:
            pass
        
        if f == 'nr':
            line_name_list[-1].append(w)  # 统计该段落的人物
            if w in name_cnt_dict.keys():  # 存在该人物则其对应value加一
                name_cnt_dict[w] = name_cnt_dict[w] + 1
            else:
                name_cnt_dict[w] = 1  # 不存在则其value为1
        # 打印进度
        count = count + 1
        count_quo = int(count/1000)
        count_mod = count % 1000 # 取模，即做除法得到的余数
        if count_mod == 0: # 每逢整千的数，打印一次进度
            print('\r已处理词数：' + '-'*count_quo + '> '\
                  + str(count_quo) + '千', end='')  # 自行刷新的进度条

relation_dict = {}  # 统计共现关系的字典

# 确定统计门槛
print()
name_cnt_limit = int(input('请输入需要统计的人物名词出现最低次数：'))
link_cnt_limit = int(input('请输入需要统计的共现最低次数：'))

for line_name in line_name_list:  # 分段处理
    for name1 in line_name:
        # 判断该人物name1是否在字典中
        if name1 in relation_dict.keys():
            pass
        elif name_cnt_dict[name1] >= name_cnt_limit:  # 统计次数大于出现次数最低要求的人名
            relation_dict[name1] = {}  # 创建该人名的字典
        else:
            continue
        
        # 统计name1与其他人名的共现数量
        for name2 in line_name:
            # 不统计name1和不满足最低出现次数的人名
            if name2 == name1 or name_cnt_dict[name2] < name_cnt_limit:  
                continue
            # 统计该段中共现次数
            if name2 in relation_dict[name1].keys():  # 如果存在则其value加一
                relation_dict[name1][name2] = relation_dict[name1][name2] + 1
            else:
                relation_dict[name1][name2] = 1  # 不存在则其value为1

# 写入csv文件
link_file = open('result3.csv', 'w')  # 打开文件
link_file.write('Source,Target,Weight\n')
for name1, link_dict in relation_dict.items():

    # 对该和该人名共现的人名的共现次数进行排序
    item_list = list(link_dict.items())
    item_list.sort(key=lambda x:x[1],reverse=True)
    
    for name2, link in item_list:
        # 写入共现次数大于最低要求的人名及其共现次数
        if link >= link_cnt_limit:
            link_file.write(name1 + ',' + name2 + ',' + str(link) + '\n')
link_file.close()

print('已成功写入result3.csv')
```
